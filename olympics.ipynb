```python
import pandas as pd
```

```python
bios=pd.read_csv('/Users/kseniaoblasova/Downloads/bios.csv')
bios
```

```python
bios.drop(columns=bios.filter(like='date'), inplace=True)
bios
```

```python
bios.filter(regex=r'^\w+_\w+$')
```

```python
bios.rename(columns={'athlete_id':'id'}, inplace=True)
bios
```

```python
bios.columns=bios.columns.str.replace('_area', '_region')
bios
```

```python
del bios['double_weight_kg']
bios
```

```python
bios=bios.add_prefix('athlete_')
bios
```

```python
bios.columns=bios.columns.str.replace('athlete_', '')
bios
```

```python
del bios['id']
bios
```

```python
bios.insert(loc=0, column='athlete_id', value=range(1, len(bios)+1))
bios
```

```python
bios.born_city=bios['born_city'].str.replace('Moskva (Moscow)', 'Moscow')
bios.born_region=bios['born_region'].str.replace('Moskva', 'Moscow')
bios[bios['born_region']=='Moscow'].sample(20)
```

```python
bios['NOC']=bios['NOC'].replace({'Soviet Union': 'USSR'})
bios[bios['born_region']=='Moscow'].sample(20)
```

```python
mask=bios['NOC'].str.contains('Russian')
contains_russian=bios[mask]
contains_russian.sample(20)
```

# Styling

```python
sns.set_style('ticks')

plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale=1.5)
sns.jointplot(x='speeding',y='alcohol',data=df,kind='reg')

sns.despine(left=True,bottom=True)
```

# Categorical Plots

# Bar Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

# Count Plot

```python
sns.countplot(x='sex',data=tips)
```

# Box Plot

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
plt.legend(loc=0)
```

# Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# Strip Plot

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day',y='total_bill',data=tips,
              jitter=True,hue='sex',dodge=True)
```

# Swarm Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

# Palettes

```python
plt.figure(figsize=(8,5))

sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',
              palette='afmhot')

plt.legend(loc=2)
```

# Pair Grids

# Facet Grid

```python
tips_fg=sns.FacetGrid(tips, col='time',
                      row='smoker')

tips_fg.map(plt.hist,'total_bill',bins=8)
```

```python
kws=dict(s=50,linewidth=0.5, edgecolor='w')
tips_fg=sns.FacetGrid(tips,col='sex',hue='smoker',
                      height=4, aspect=1.3,
                      hue_order=['Yes','No'],
                      hue_kws=dict(marker=['^','v']))
tips_fg.map(plt.scatter, 'total_bill','tip',**kws)
```

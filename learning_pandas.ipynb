# Series

```python
data=[100,102,104, 200, 202]
series=pd.Series(data, index=['a','b','c', 'd', 'e'])
series.loc['c']=105
print(series[series<200])
```

```python
calories={'Day 1':1750, 'Day 2': 2100, 'Day 3': 1700}
series_a=pd.Series(calories)
series_a.loc['Day 3']+=300
print(series_a[series_a<2000])
```

# DataFrames

```python
data={'Name':['Kseniia', 'Iuliia', 'Aleksandr'],
      'Age':[18,52,53]}

df=pd.DataFrame(data, index=['Fam member 1', 'Fam member 2', 'Fam member 3' ])
#add a new column

df['Job']=['Student','N/A','Constructor']

#Add a new row
new_rows=pd.DataFrame([{'Name':'Yana', 'Age':29, 'Job': 'Cosmetologist'},
                      {'Name':'Oleksandr', 'Age':35, 'Job': 'N/A'}], 
                     index=['Fam member 4', 'Fam member 5'])

df=pd.concat([df, new_rows])


print(df)
```

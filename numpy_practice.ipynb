# Practice Problem 1

```python
rng=np.random.default_rng(seed=33)
some_data=rng.integers(100,450,size=(100,4))
print(some_data)
```

```python
mean_val=np.mean(some_data[:,0])
std_val=np.std(some_data[:,0])
print(mean_val)
print(std_val)
```

```python
lower_bound=mean_val-(2*std_val)
upper_bound=mean_val+(2*std_val)
mask=((some_data[:,0]>lower_bound)&(some_data[:,0]<upper_bound))
#or
#mask=np.abs(some_data[:,0]-mean_val)<(2*std_val)
some_data=some_data[mask]
print(some_data)
```

# Practice Problem 2

```python
rng=np.random.default_rng(seed=42)
ages=rng.integers(0,101, size=(10,1))
incomes=rng.integers(20000,80000,size=(10,1))

X=np.hstack((ages, incomes))
print(ages)

print(X)
```

```python
min_vals=X.min(axis=0)
max_vals=X.max(axis=0)

print(min_vals)
print(max_vals)
```

```python
numerator=X-min_vals
denominator=max_vals-min_vals
scaled_data=numerator/denominator
print(scaled_data)

```

# Practice Problem 4

```python
rng=np.random.default_rng(seed=42)
grades=rng.integers(0,101,(10,5))
print(grades)
```

```python
mean_scores=grades.mean(axis=1)
max_scores=grades.max(axis=1)
print(mean_scores.reshape(-1,1))
print(max_scores)
```

```python
grades_stats=np.hstack((grades, mean_scores.reshape(-1,1), max_scores.reshape(-1,1)))
print(grades_stats)

```
